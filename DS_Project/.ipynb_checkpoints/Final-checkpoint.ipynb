```python
import pandas as pd

url = "https://shazam.p.rapidapi.com/charts/track"
headers = {
    "X-RapidAPI-Key": "13ab037ab5msh5e81139da3210adp1c1d7fjsn72a711f8713d",
	"X-RapidAPI-Host": "shazam.p.rapidapi.com"
}

records = []
page_size = 20
total_records = 100 #number of records to collect
start_from = 0

while len(records) < total_records:
    params = {"locale":"en-US",
              "listId":"genre-global-chart-1", # = POP
              "pageSize":str(page_size),
              "startFrom":str(start_from)
             }
    """
    defining specific parameters
    """

    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        data = response.json().get('tracks')
        records.extend(data)
        start_from += page_size
    else:
        print("Error:", response.status_code)
        break
    """
    transforming JSON into a pandas DataFrame
    using a loop to increment the 'startFrom' parameter in each request
    until it accumulates a chart of Top 100
    """

pop_df = pd.DataFrame(records)
print(pop_df.head())
```

```python
import pandas as pd

url = "https://shazam.p.rapidapi.com/charts/track"
headers = {
    "X-RapidAPI-Key": "13ab037ab5msh5e81139da3210adp1c1d7fjsn72a711f8713d",
	"X-RapidAPI-Host": "shazam.p.rapidapi.com"
}

records = []
page_size = 20
total_records = 100
start_from = 0

while len(records) < total_records:
    params = {"locale":"en-US",
              "listId":"genre-global-chart-10", # = Country
              "pageSize":str(page_size),
              "startFrom":str(start_from)
             }
    """
    defining specific parameters
    """

    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        data = response.json().get('tracks')
        records.extend(data)
        start_from += page_size
    else:
        print("Error:", response.status_code)
        break
    """
    transforming JSON into a pandas DataFrame
    using a loop to increment the 'startFrom' parameter in each request
    until it accumulates a chart of Top 100
    """

country_df = pd.DataFrame(records)
print(country_df.head())
```

```python
pop_df.rename(columns={'subtitle':'artist'}, inplace=True) 
country_df.rename(columns={'subtitle':'artist'}, inplace=True)
# inplace=True applies the changes directly to the DataFrame without reassigning it to a new var.
"""
renaming 'subtitle' column to 'artist'
"""
print(pop_df.head())
print(country_df.head())
```

```python
def match_letter_cases(df, columns_match, case = 'lower'):
    for col in columns_match:
        if col in df.columns:
            if case == 'lower':
                df[col] = df[col].apply(lambda x: x.lower() if isinstance(x, str) else x)
            elif case == 'upper':
                df[col] = df[col].apply(lambda x: x.upper() if isinstance(x, str) else x)
            elif case == 'title': #capitalizing the first letter of each word
                df[col] = df[col].apply(lambda x: x.title() if isinstance(x, str) else x)
            else:
                raise ValueError("Invalid case.")
    return df
    """
    standardizing letter cases
    """
```

```python
columns_match = ['title', 'artist']
pop_transformed = match_letter_cases(pop_df, columns_match, case = 'title')
country_transformed = match_letter_cases(country_df, columns_match, case = 'title')
print(pop_transformed.head())
print(country_transformed.head())
```

```python
pop_transformed.to_csv('pop_charts.csv', index=False)
country_transformed.to_csv('country_charts.csv', index=False)
"""
saving as .csv files
"""
```
