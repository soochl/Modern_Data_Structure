# Selection of the Data and Reshaping

```python
df.drop(df.columns.difference(['YearEnd', 'LocationAbbr', 'LocationDesc', 'Question', 'DataValueType', 'DataValue', 'Stratification1']), axis=1, inplace=True)
df
```

```python
df = df.loc[~((df['Question'] == 'Poverty') & (df['Stratification1'] == 'Male'))]
df = df.loc[~((df['Question'] == 'Poverty') & (df['Stratification1'] == 'Female'))]
df
```

```python
df_filtered = df[df['Question'].isin(['Poverty' , 'Binge drinking prevalence among adults aged >= 18 years']) & df['DataValueType'].isin(['Crude Prevalence']) & df['Stratification1'].isin(['Overall', 'Male', 'Female'])]
df_filtered
```

```python
df_wide = df_filtered.drop_duplicates(subset=['YearEnd', 'Question'])
df_wide
```

```python
df_rename = df_wide.rename(columns={'YearEnd':'year','LocationDesc':'state', 'LocationAbbr':'stateabb'})
df_rename
```

```python
df_cleaned = df_rename.pivot(index=['year','stateabb', 'state'], columns= ['Question', 'Stratification1'], values ='DataValue')
df_cleaned
```

```python
df_final = df_cleaned.rename(columns={'Overall': 'all', 'Male':'binge_male', 'Female':'binge_female'})
df_final
```

# Data Transformation and Summary Results

```python
binge_df = df[df['Question'].isin(['Binge drinking prevalence among adults aged >= 18 years']) & df['Stratification1'].isin(['Overall', 'Male', 'Female'])]
binge_df
```

```python
binge_filtered = binge_df.drop_duplicates(subset=['LocationDesc'])
binge_filtered
```

```python
binge_rename = binge_filtered.rename(columns={'YearEnd':'year','LocationDesc':'state', 'LocationAbbr':'stateabb'})
binge_rename
```

```python
binge_wide = binge_rename.pivot(index=['year','stateabb', 'state'], columns= ['Stratification1'], values ='DataValue')
binge_wide
```

```python
binge_wide[["Female"]] = binge_wide[["Female"]].apply(pd.to_numeric)
binge_wide[["Male"]] = binge_wide[["Male"]].apply(pd.to_numeric)
binge_wide[["Overall"]] = binge_wide[["Overall"]].apply(pd.to_numeric)
```

```python
binge_top_ten = binge_wide.sort_values(by='Overall', ascending=False)
binge_top_ten.head(10)
```

```python
binge_rename[["DataValue"]] = binge_rename[["DataValue"]].apply(pd.to_numeric)
```

```python
binge_gr = binge_rename[['year', 'state', 'DataValue']].groupby('state').first()
binge_gr
```

```python
binge_gr['Growth_Rate'] = binge_gr['DataValue'].pct_change(periods=1) * 100
binge_gr
```

```python
binge_gr_filtered = binge_gr.dropna()
binge_gr_filtered
```

```python
binge_increase = binge_gr_filtered.nlargest(5, 'DataValue', keep = 'all')
binge_decrease = binge_gr_filtered.nsmallest(5, 'DataValue', keep = 'all')
```

```python
pd.concat([binge_increase, binge_decrease], ignore_index=False)
```
