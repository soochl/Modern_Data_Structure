# a) Length of narratives

```python
df.head()
```

```python
def narrative_len(narrative):
    return len(re.sub(r'\W+', '', narrative))
```

```python
df['length'] = df['narrative'].apply(narrative_len)
```

```python
longest_narrative = df.loc[df['length'].idxmax()]
shortest_narrative = df.loc[df['length'].idxmin()]
```

```python
print('Longest narrative: {} characters'.format(longest_narrative['length']))
print('First 50 characters:', longest_narrative['narrative'][:50])
```

```python
print('\nShortest narrative: {} characters:'.format(shortest_narrative['length']))
print('First 5 characters:', shortest_narrative['narrative'][:5])
```

# b) Fractures

```python
narratives = ' '.join(df['narrative'].astype(str))
```

```python
fracture_regex = r'\bfractur(?:e|es|ed)?\b|FX'
```

```python
fracture_match = re.findall(fracture_regex, narratives, re.IGNORECASE)
```

```python
fracture = len(fracture_match)
print(fracture)
```

# c) Body parts

```python
body_type = r'\b(head|neck|arm|hand|finger|leg|ankle|foot|toe|face|nose|lip|eye|back)\b'
```

```python
body_parts = re.findall(body_type, narratives, re.IGNORECASE)
print(body_parts)
```

```python
body_part_occ = {body_part: body_parts.count(body_part) for body_part in set(body_parts)}
```

```python
print (body_part_occ)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
plt.bar(body_part_occ.keys(), body_part_occ.values())
plt.xlabel('Body Parts')
plt.ylabel('Occurence(s)')
plt.title('Count of Fractured Body Parts')
plt.tight_layout()
plt.show()
```

# d) Age and Gender

```python
age_gender_info = df['narrative']
```

```python
age_regex = r'(\d\d)s*(YO|MO)'
gender_regex = r'([M|F])'
```

```python
age_gender_regex = r'(\d\d)s*(YO|MO)?(M|F)?'
age_conv = {'YO':1, 'MO': 1/12}
```

```python
ages = []
genders = []
```

```python
for narrative in age_gender_info:
    age_gender = re.findall(age_gender_regex, narrative)
    age = 0
    for i in age_gender:
        num, unit, gender = i
        age += int(num) * age_conv.get(unit.upper(), 1)

    ages.append(age)
    genders.append(gender)
```

```python
df['age'] = ages
df['gender'] = genders
```

```python
print(df[['narrative', 'age', 'gender']])
```

```python
age_gender_regex = r'(\d\d)s*(YO|MO)?(M|F)?'
age_conv = {'YO':1, 'MO': 1/12}
```

```python
ages = []
genders = []
```

```python
for narrative in age_gender_info:
    age_gender = re.findall(age_gender_regex, narrative)
    
    if age_gender:
        age = 0
        gender = 'Unknown'

        for i in age_gender:
            num, unit, g = i
            if unit:
                age += int(num) * age_conv.get(unit.upper(), 1)

            if g:
                gender = g
        ages.append(age if age >0 else 'Not specified')
        genders.append(gender)
    else:
        ages.append('Not specified')
        genders.append('Unknown')
```

```python
df['age'] = ages
df['gender'] = genders
```

```python
print(df[['narrative', 'age', 'gender']])
```

```python
def age_groups(age):
    if age == 'Not specified':
        return 'Not specified'
    elif age < 20:
        return '0-19'
    elif age < 30:
        return '20-29'
    elif age < 40:
        return '30-39'
    elif age < 50:
        return '40-49'
    else:
        return '50+'

df['age_group'] = df['age'].apply(age_groups)
    
```

```python
age_gender_injury = df.groupby(['age_group', 'gender']).size().unstack()
```

```python
plt.figure(figsize=(10, 5))
age_gender_injury.plot(kind='bar', stacked = False)
plt.xlabel('Age Group')
plt.ylabel('Number of Injuries')
plt.title('Number of Injuries by Age Group')
plt.legend(title='Gender', labels=['Female', 'Male', 'Unknown'])
plt.tight_layout()
plt.show()
```
