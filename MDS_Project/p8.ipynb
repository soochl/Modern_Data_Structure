```python
def extract_infobox(html_content):
    bs = BeautifulSoup(html_content, 'html.parser')
    infobox = bs.find('table', {'class': 'infobox'})
    
    if infobox:
        return infobox

def infobox_attribute(infobox, attribute_name):
    if infobox:
        attribute = infobox.find('th', {'class': 'infobox-label'}, string=attribute_name)
        if attribute:
            value = attribute.find_next('td')
            if value:
                return value.get_text(strip=True)
    alt_attribute_names = {
        "Gross revenue": "Box office"
    }
    for alt_name, alt_value in alt_attribute_names.items():
        if alt_name.lower() == attribute_name.lower():
            return extract_infobox_attribute(infobox, alt_value)

    print(f"Attribute '{attribute_name}' not found in the content.")
    return None
```

```python
def infobox_exceptions(attribute_name, attribute_value):
    if attribute_value is None:
         print(f"Information for attribute '{attribute_name}' is missing.")
    else:
        return attribute_value
```

```python
url = "https://en.wikipedia.org/wiki/The_Shape_of_Water"
html_content = get_html_content(url)

if html_content:
    infobox = extract_infobox(html_content)
    attribute_name = "Box office"
    attribute_value = infobox_attribute(infobox, attribute_name)

    handled_except = infobox_exceptions(attribute_name, attribute_value)
    if handled_except:
        print(f"{attribute_name}: {attribute_value}")
```

# 03. Structured DataFrame

```python
def get_html_content(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        print(f"Failed to retrieve the HTML content.")
        return None
        
def extract_infobox_content(html_content):
    bs = BeautifulSoup(html_content, 'html.parser')
    infobox = bs.find('table', {'class': 'infobox'})
    if infobox:
        infobox_content = []
        for row in infobox.find_all('tr'):
            header = row.find('th')
            data = row.find('td')

            if header and data:
                attribute = header.get_text(strip=True)
                value = data.get_text(strip=True)
                infobox_content.append({'attribute': attribute, 'value': value})
        return pd.DataFrame(infobox_content)
    else:
        print("Error: Infobox not found on the Wikipedia page.")
        return None 
```

```python
def running_time(time_str):
    # Convert running time to minutes
    try:
        parts = time_str.split()
        minutes = 0
        for part in parts:
            if 'h' in part:
                minutes += int(part.replace('h', '')) * 60
            elif 'min' in part:
                minutes += int(part.replace('min', ''))
        return minutes
    except ValueError:
        return None

def currency(currency_str):
    # Normalize currency format
    try:
        value = float(currency_str.split()[0].replace(',', '').replace('$', '').replace('million', '')) * 1e6
        return value
    except ValueError:
        return None

def clean_data(df):
    df.loc[df['attribute'] == 'Running time', 'value'] = df.loc[df['attribute'] == 'Running time', 'value'].apply(running_time)
    df.loc[df['attribute'] == 'Box office', 'value'] = df.loc[df['attribute'] == 'Box office', 'value'].apply(currency)
    
    return df
```

```python
url = "https://en.wikipedia.org/wiki/The_Shape_of_Water"
html_content = get_html_content(url)

if html_content:
    infobox_data = extract_infobox_content(html_content)
    if infobox_data is not None:
        infobox_data_cleaned = clean_data(infobox_data)

print("Cleaned DataFrame:")
print(infobox_data_cleaned)

infobox_data_cleaned.to_csv("infobox_data.csv", index=False)
print("DataFrame exported to infobox_data.csv")    
```

# 04. Robust Scraping Function

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_oscar_movie_info(url):
    try:
        response = requests.get(url)
        bs = BeautifulSoup(response.content, 'html.parser')
        infobox = bs.find('table', {'class': 'infobox'})
        if not infobox:
            raise ValueError("Infobox not found in the page")
 
        rows = infobox.find_all('tr')
        data = []

        for row in rows:
            if row.find('th') and row.find('td'):
                attribute = row.find('th').text.strip()
                value = row.find('td').text.strip()
                data.append([attribute, value])

        df = pd.DataFrame(data, columns=['attribute', 'value'])

        csv_filename = url.split("/")[-1] + "_data.csv"
        df.to_csv(csv_filename, index=False)
        print(f"Generated {csv_filename}")

        return df
    except Exception as e:
        print(f"Error")
        return pd.DataFrame()
```

```python
url = 'https://en.wikipedia.org/wiki/Moonlight_(2016_film)'
moonlight_df = get_oscar_movie_info(url)
print(moonlight_df)
```

```python
url = 'https://en.wikipedia.org/wiki/Parasite_(2019_film)'
parasite_df = get_oscar_movie_info(url)
print(parasite_df)
```

# 05. Analyze All Best Picture Winners

```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_best_picture_winners_urls():
    url = "https://en.wikipedia.org/wiki/List_of_Academy_Award-winning_films"
    response = requests.get(url)
    if response.status_code != 200:
        raise ValueError
        print(f"Attribute '{attribute_name}' not found in the content.")
    html_content = response.content   
    bs = BeautifulSoup(html_content, 'html.parser')

    winners_list = []
    winners_table = bs.find('table',{'class': 'wikitable sortable'})
    rows = winners_table.find_all("tr", {"style": "background:#EEDD82"})
        
    for row in rows:
        cells = row.find_all('td')
        if len(cells) > 1:
            winner_link = cells[0].find('a')
            if winner_link and 'href' in winner_link.attrs:
                winners_list.append("https://en.wikipedia.org" + winner_link['href'])
    return winners_list

best_picture_winner_urls = get_best_picture_winners_urls()
print (len(best_picture_winner_urls))
```

```python
best_picture_urls = get_best_picture_winners_urls()
all_films_data = {}

for url in best_picture_urls:
    film_name = url.split("/")[-1]
    print(f"Processing: {film_name}")

    try:
        film_data = get_oscar_movie_info(url)
        all_films_data[film_name] = film_data
    except Exception as e:
        print(f"Failed to retrieve data for {film_name}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfs = []

for film_name in all_films_data.keys():
    csv_filename = film_name + "_data.csv"
    try:
        df = pd.read_csv(csv_filename)
        df['Film'] = film_name
        dfs.append(df)
    except Exception as e:
        print(f"Error reading {csv_filename}: {e}")

best_picture_df = pd.concat(dfs, ignore_index=True)

best_pic_csv_filename = "best_picture_data_all.csv"
best_picture_df.to_csv(best_pic_csv_filename, index=False)
```

```python
data_path = 'best_picture_data_all.csv'
best_picture_df = pd.read_csv(data_path)

top_distributor = best_picture_df[best_picture_df['attribute'] == 'Distributed by']['value'].mode()[0]
print(f"The distributor with the most Best Picture winners is: {top_distributor}")
```

```python
data_path = 'best_picture_data_all.csv'
length_data = pd.read_csv(data_path)

running_times = length_data[length_data['attribute'] == 'Running time']
running_times['value'] = running_times['value'].str.extract('(\\d+)').astype(int)

plt.figure(figsize=(12, 6))
plt.bar(running_times['Film'], running_times['value'], color='blue')
plt.title('Length of Best Picture Winning Films Over Time')
plt.xlabel('Film')
plt.xticks(rotation=150)
plt.ylabel('Running Time (minutes)')
plt.show()
```
